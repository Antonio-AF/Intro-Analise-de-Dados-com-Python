```python
# carregar o arquivo com o resultado da eleição no brasil dividido por município/zona.

df_votacao = pd.read_csv('./dados_eleicao/votacao_candidato_munzona_2018_BR.csv',
        sep=';', encoding='latin1')


df_votacao.head()
```

```python
# Visualizando o DataFrame transposto

df_votacao.head().T
```

```python
# Separar os dados pois precisamos dos dados de segundo turno


df_votacao = df_votacao[df_votacao['NR_TURNO'] == 2]

df_votacao.sample(3).T
```

```python
# Vamos verificar quantos municipios unicos exitem no conjunto de dados
# Vamos pegar por codigo do município

df_votacao['CD_MUNICIPIO'].nunique()
```

```python
# Mas o brasil tem 5570 municípios
# Vamos verificar algumas cidades fora do Brasil.

df_votacao[df_votacao['SG_UF'] == 'ZZ'].sample(4).T
```

```python
# Quantas cidades fora do Brasil estõ no conjunto de dados.

df_votacao[df_votacao['SG_UF'] == 'ZZ'] ['CD_MUNICIPIO'].nunique()
```

```python
# Vamos eliminar os municipios fora do Brasil


df_votacao = df_votacao[df_votacao['SG_UF'] != 'ZZ']
```

```python
# Verificar a quantidade de município

df_votacao['CD_MUNICIPIO'].nunique()
```

```python
# Os dados estão separados por Municípios/Zona.
# Municípios pequenos normalmente tem apenas 1 zona eleitoral mas, 
# municípios maiopres podem ter várias zonas eleitorais.


df_votacao[df_votacao['NM_MUNICIPIO'] == 'BELO HORIZONTE']
```

```python
# Eliminar algumas colunas do DdataFrame que não seram utilizadas

df_votacao = df_votacao[['SG_UF', 'CD_MUNICIPIO', 'NM_MUNICIPIO', 'NR_CANDIDATO',
                         'NM_URNA_CANDIDATO', 'QT_VOTOS_NOMINAIS']]

df_votacao.head(20)
```

```python
# Precisamos agrupar os dados por município e candidado.

df_votacao.groupby(['SG_UF', 'CD_MUNICIPIO', 'NM_MUNICIPIO', 'NR_CANDIDATO',
                    'NM_URNA_CANDIDATO']).sum()

```

```python
# Fazer o groupby, reindexar o indice, e salvar  isso em um DataFrame.

df_votacao = df_votacao.groupby(['SG_UF', 'CD_MUNICIPIO', 'NM_MUNICIPIO', 'NR_CANDIDATO',
                    'NM_URNA_CANDIDATO']).sum().reset_index()

df_votacao.sample(4)
```

```python
# Vamos ordenar o Dataframe pela quantidade de votos e depois deixar apenas o primeiro
# registro de cada município.

df_votacao.sort_values(by='QT_VOTOS_NOMINAIS', ascending=False).head(10)
```

```python
# Vamos ordenar 
df_votacao.sort_values(by='QT_VOTOS_NOMINAIS', ascending=False, inplace=True)


# Dropar as linhas duplicadas, mantendo a primeira
df_votacao.drop_duplicates(subset='CD_MUNICIPIO', keep='first', inplace=True)

df_votacao.head(10)
```

```python
# Agora vamos criar uma coluna representando o resultado da eleição naquele municipio.
# Sendo o resultado "True" se o candidato Bolsonaro venceu naquele municipio e "False" se caso contrario.

df_votacao['RESULTADO'] = df_votacao['NR_CANDIDATO'].apply(lambda x: x == 17)
```

```python
# Verificando o resultado 

df_votacao.head(10)
```

```python
# Mostrar alguns valores do DataFrame do geopandas

df_mapa.sample(3)
```

```python
# Tentando encontrar o município pelo código.

df_votacao[df_votacao['CD_MUNICIPIO'] == 3146255]
```

```python
# Porque não podemos usar o nome do município para juntar.
# Pois temos alguns Municipios com o mesmo nome, em diferentes estados.
df_votacao[df_votacao['NM_MUNICIPIO'] == 'OURO BRANCO']
```

```python
# Vamos recuperar a tabela de correspondencia entre IBGE e TSE
# direto do github do Estadão

link = 'https://raw.githubusercontent.com/estadao/como-votou-sua-vizinhanca/master/data/votos/correspondencia-tse-ibge.csv'

df_equivalencia =pd.read_csv(link)

df_equivalencia.head()
```

```python
# Vamos reindexar o DataFrame de votação pelo código do Municipio


df_vot_novo = df_votacao.copy()

df_vot_novo.set_index('CD_MUNICIPIO', drop=False, inplace=True)

df_vot_novo.head()
```

```python
# Vamos reindexar o DataFrame de equivancia pelo código do TSE

df_equi_novo = df_equivalencia.copy()

df_equi_novo.set_index('COD_TSE', drop=False, inplace=True)

df_equi_novo.head()
```

```python
# Verificando o tamanho dos DataFrame


df_vot_novo.info()
```

```python
df_equi_novo.info()
```

```python
# Criando um novo DataFrame com base no merge de Votação e equivalencia

df_vot_equi = df_vot_novo.join(df_equi_novo)

df_vot_equi.head()
```

```python
# Vamos juntar os Dataframes do Geopandas com o DataFrame Votacao/Equivalencia.

```

```python
df_vot_equi.info()
```

```python
df_mapa.info()
```

```python
# transformar a coluna de código de IBGE do dataframe de Votação/Equivalencia


df_vot_equi['GEOCOD_IBGE'] = df_vot_equi['GEOCOD_IBGE'].astype(str)

df_vot_equi.info()
```

```python
# Vamos Unificar os Indices


# Mudando o indice do dataframe de Mapa

df_mapa.set_index('CD_GEOCMU', drop=False, inplace=True)

```

```python
# Vamos unicficar o dataframe do Mapa com o dataframe de Votação/Equivalencia

df_vot_equi.set_index('GEOCOD_IBGE', drop=False, inplace=True)
```

```python
# Vamos juntar os Dataframe.


df_mapa_novo = df_mapa.join(df_vot_equi)


df_mapa_novo.head().T
```

```python
# Vamos verificar o tamanho do Mapa Novo

df_mapa_novo.info()
```

```python
# Vamos descobrir os dois Municípios a mais no mapa


df_mapa_novo[df_mapa_novo['NR_CANDIDATO'].isnull()]
```

```python
# Vamos Matar a coluna Ajuste

df_mapa_novo.drop(columns=['AJUSTE'], inplace=True)
```

```python
# Vamos eliminar a informação das duas lagoa do RS.

df_mapa_novo.dropna(inplace=True)
```

```python
df_mapa_novo.info()
```

```python
# Vamos desenhar o mapa.


# Criar a referencia para o código de cor (False/0 = red, True/1 = green)


from matplotlib.colors import ListedColormap

cmap = ListedColormap(['red', 'green'])

# Desenhar o Mapa
fig, ax = plt.subplots(1, figsize=(12, 12))


ax = df_mapa_novo.plot(column ='RESULTADO', cmap=cmap, legend=True, linewidth=1, ax=ax,
                edgecolor='grey')


# Esconder os Eixos.

ax.set_axis_off()

# Vamos colocar um Titulo no Mapa

ax.set_title('Resultado Elição 2018 por Município / 2º Turno', fontdict={'fontsize':'25',
             'fontweight':'3'})

             
# Criando uma nota do rodapé.
             
ax.annotate('Fonte: IBGE, 2015; TSE, 20199; Jornal o Estado de São Paulo, 2019',
             xy=(0.1, 0.08), xycoords='figure fraction', horizontalalignment='left',
             verticalalignment='top', fontsize=10, color='#555555')
             
             
# Alterando a Legenda.            
leg = ax.get_legend()

leg.get_texts()[0].set_text('Fernando Haddad')
leg.get_texts()[1].set_text('Jair Bolsonaro')
             
             
plt.show()             
```
